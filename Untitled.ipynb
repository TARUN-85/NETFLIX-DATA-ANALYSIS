### Preprocessing Of Data

#### 1.Conversion of duration in float dtype

```python
df['duration'] = pd.to_numeric(df['duration'].str.split(' ').str[0])
```

#### 2.Converting data into Datetime format

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

#### 3.1 Splitting Nested Cast Data

```python
df['cast'] = df['cast'].str.split(',')
df = df.explode('cast')
```

#### 3.2 Splitting Nested Genre Data

```python
df['listed_in'] = df['listed_in'].str.split(',')
df = df.explode('listed_in')
```

#### 3.3 Splitting Nested Country Data

```python
df['country'] = df['country'].str.split(',')
df = df.explode('country')
```

```python
df['country'].replace(' United Arab Emirates','United Arab Emirates',inplace=True)
```

#### 4.Filling Missing Value

> Missing value for categorical variable can be filled by taking modal value(The most occuring value) along a group by to maintain homogeniety od director & cast

##### Filling of duration column

```python
df['duration'].isna().sum()
```

```python
df['duration'].fillna(df['rating'].str.split(' ').str[0],inplace=True)
```

```python
df['duration'].isna().sum()
```

#### Filling of cast

```python
#Replacing Modal value for a particular director & cast combination
```

```python
df['cast'].isna().sum()
```

```python
a = df.groupby(['director'])['cast'].agg(pd.Series.mode).explode().reset_index()
df['cast'].fillna(a['cast'],inplace=True)
```

```python
df['cast'].isna().sum()
```

```python
# it is around % & we dont have information abouth this cast so either we can drop this row or proceed further..
#its 5 better to drop it
```

```python
df.dropna(how='any',subset='cast',inplace=True)
df['cast'].isna().sum()
```

#### Filling of director

```python
df['director'].isna().sum()
```

```python
c = df.groupby(['cast'])['director'].agg(pd.Series.mode).explode().reset_index()
df['director'].fillna(c['director'],inplace=True)
```

```python
#after filling null vaue
df['director'].isna().sum()
```

```python
#dropping the remining null value as ther we negligible & do not affect the op much as less no of qty
df.dropna(how='any',subset='director',inplace=True)
```

#### Filling of Country

```python
df['country'].isna().sum()
```

```python
cntry = df.groupby(['cast','director'])['country'].agg(pd.Series.mode).explode().reset_index()
df['country'].fillna(cntry['country'],inplace=True)
```

```python
df['country'].isna().sum()
```

```python
df.dropna(how='any',subset='country',inplace=True)
```

#### Filling of Ratings

```python
df['rating'].isna().sum()
```

```python
rtng = df.groupby(['cast','director','country'])['rating'].agg(pd.Series.mode).explode().reset_index()
df['rating'].fillna(rtng['rating'],inplace=True)
```

```python
df['rating'].isna().sum()
```
